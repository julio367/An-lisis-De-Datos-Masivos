```python
df_incidentes = pd.read_csv('Police.csv')
df_incidentes.head()
```

```python
df_incidentes.shape
```

```python
limite = 100
df_incidentes = df_incidentes.iloc[0:limite, :]
```

```python
df_incidentes.shape
```

```python
latitude = 37.77
longitude = -122.42
sf_map = folium.Map(location = [latitude,longitude], zoom_start=12)
sf_map
```

```python
incidentes = folium.map.FeatureGroup()
for lat, lng in zip(df_incidentes.Y, df_incidentes.X):
    incidentes.add_child(
    folium.CircleMarker(
    [lat,lng], radius=5, color='yellow', fill=True, fill_color='blue',fill_opacity=0.6)
    )
sf_map.add_child(incidentes)
```

#### Colocar texto emergente sobre los marcadores

```python
incidentes = folium.map.FeatureGroup()
for lat, lng in zip(df_incidentes.Y, df_incidentes.X):
    incidentes.add_child(
    folium.CircleMarker(
    [lat,lng], radius=5, color='yellow', fill=True, fill_color='blue',fill_opacity=0.6)
    )

latitudes = list(df_incidentes.Y)
longitudes = list(df_incidentes.X)
labels = list(df_incidentes.Category)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(sf_map)
    
sf_map.add_child(incidentes)
```

```python
latitude = 37.77
longitude = -122.42
sf_map = folium.Map(location = [latitude,longitude], zoom_start=12)
for lat, lng, label in zip(df_incidentes.Y, df_incidentes.X, df_incidentes.Category):
    folium.CircleMarker(
    [lat,lng], radius=5, color='yellow', fill=True, fill_color='blue',fill_opacity=0.6, popup=label).add_to(sf_map)
    
sf_map
```

```python
from folium import plugins
latitude = 37.77
longitude = -122.42
sf_map = folium.Map(location = [latitude,longitude], zoom_start=12)
incidentes = plugins.MarkerCluster().add_to(sf_map)
for lat, log, label in zip(df_incidentes.Y, df_incidentes.X, df_incidentes.Category):
    folium.Marker(
    location=[lat,lng],
    icon=None,
    popup=label).add_to(incidentes)
    
sf_map
```
